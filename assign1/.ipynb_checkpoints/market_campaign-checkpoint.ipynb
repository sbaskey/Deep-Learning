```python
import pandas as pd
```

```python
#load data set
data = pd.read_excel('marketing_campaign.xlsx')
```

```python
data.shape
```

```python
data.describe()
```

```python
#split the data set into input and output set
x=data.drop(["Offer_acceptance"],axis=1)
y=data["Offer_acceptance"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#sampling the the data set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3)
```

```python
from sklearn import tree
```

```python
ctree=tree.DecisionTreeClassifier()
```

```python
ctree=ctree.fit(x_train,y_train)
```

```python
y_pred=ctree.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn import metrics
```

```python
#default model
ctree = tree.DecisionTreeClassifier()
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
ctree_model = ctree.fit(x_train,y_train)
y_train_pred = ctree_model.predict(x_train)
print('train_accuracy:',metrics.accuracy_score(y_train,y_train_pred))
y_test_pred = ctree_model.predict(x_test)
print('test_accuracy:',metrics.accuracy_score(y_test,y_test_pred))
```

```python
#comparing actual vs real
comparision_df = x_test.copy()
comparision_df['actual'] = y_test
comparision_df['Prediction'] = y_test_pred
comparision_df
```

```python
from sklearn.model_selection import cross_val_score
```

```python
#cross_validation
cross_val_score(ctree,x,y,cv=5,scoring='accuracy')
```

```python
cross_val_score(ctree,x,y,cv=5,scoring='accuracy').mean()
```

```python
#parameter tunning
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { "criterion"       : ['gini', 'entropy'],
                "max_features"     : [4,5,6],
               "splitter"         : ['best', 'random'],
               "max_depth"        : [2,3],
               "min_samples_leaf" : [2,3]
             }
```

```python
optimized_tree = GridSearchCV(ctree, param_grid, scoring = 'accuracy', cv=5)
```

```python
optimized_tree.fit(x_train,y_train)
```

```python
optimized_tree.best_score_
```

```python
optimized_tree.best_params_
```

```python
ctree_final = optimized_tree.best_estimator_
```

```python
ctree_final
```

```python
y_pred = ctree_final.predict(x_test)
print('Accuracy', metrics.accuracy_score(y_pred, y_test))
```

```python
metrics.precision_score(y_pred,y_test)
```

```python
imp_df = pd.DataFrame()
imp_df['col_names']=x.columns
imp_df['Importance']= ctree_final.feature_importances_*100
imp_df.sort_values('Importance', ascending=False)
#imp_df
```

```python
#final model is saved
import pickle
pickle.dump(ctree_final, open('final_ctree_model_dec11.pk','wb'))
```
